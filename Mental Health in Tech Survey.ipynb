# Cleaning lanjutan & pembentukan variabel

```python
# ============================================================
# TAHAP 3 — CLEANING LANJUTAN & PEMBENTUKAN VARIABEL
# ============================================================

df_clean = df_sample.copy()

print("=== CLEANING LANJUTAN DATA ===")

# 1. Menyederhanakan Gender → Gender_simple (Male / Female / Other)

def simplify_gender(x):
    if pd.isna(x):
        return "Other"
    s = str(x).strip().lower()
    
    male_list = ["male", "m", "man", "cis male", "cis man", 
                 "male-ish", "msle", "malr", "mail", "make"]
    female_list = ["female", "f", "woman", "cis female", "cis woman", 
                   "femake", "femail"]

    if s in male_list:
        return "Male"
    elif s in female_list:
        return "Female"
    else:
        return "Other"

df_clean["Gender_simple"] = df_clean["Gender"].apply(simplify_gender)

print("Gender_simple berhasil dibuat.")

# 2. Hapus responden dengan Gender_simple = "Other"

df_clean = df_clean[df_clean["Gender_simple"] != "Other"]
print("Setelah menghapus gender 'Other':", df_clean.shape[0], "responden")
```

```python
# ------------------------------------------------------------
# 2. Membuat variabel target numerik: treatment_bin (Yes=1, No=0)
# ------------------------------------------------------------
df_clean["treatment_bin"] = df_clean["treatment"].map({"Yes": 1, "No": 0})

print("treatment_bin berhasil dibuat.")
```

```python
# ------------------------------------------------------------
# 3. Membuat kelompok umur → AgeGroup
# ------------------------------------------------------------
bins = [18, 29, 39, 49, 59, 80]
labels = ["18-29", "30-39", "40-49", "50-59", "60-80"]

df_clean["AgeGroup"] = pd.cut(df_clean["Age"], bins=bins, labels=labels, right=True)

print("AgeGroup berhasil dibuat.")
```

```python
# ------------------------------------------------------------
# 4. Merapikan kategori penting
# ------------------------------------------------------------

def clean_category(x):
    if pd.isna(x):
        return x
    s = str(x).strip().lower()
    if s in ["i don't know", "don't know", "na", "n/a"]:
        return "Unknown"
    return x

categorical_cols = [
    "benefits", "care_options", "wellness_program", "seek_help",
    "anonymity", "leave", "mental_health_consequence",
    "phys_health_consequence", "coworkers", "supervisor",
    "mental_vs_physical"
]

for col in categorical_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].apply(clean_category)

print("Kategori penting berhasil dirapikan.")
```

```python
# ------------------------------------------------------------
# 5. Menampilkan ringkasan hasil cleaning
# ------------------------------------------------------------
print("\n=== RINGKASAN SETELAH CLEANING ===")
print("Jumlah responden :", df_clean.shape[0])
print("Jumlah variabel  :", df_clean.shape[1])

print("\nContoh 5 baris pertama:")
display(df_clean.head())

print("\nFrekuensi Gender_simple:")
print(df_clean["Gender_simple"].value_counts())

print("\nFrekuensi AgeGroup:")
print(df_clean["AgeGroup"].value_counts())
```

# Analisis bivariat (chi-square)

```python
# TAHAP 6 — ANALISIS BIVARIAT (UJI CHI-SQUARE)

from scipy.stats import chi2_contingency

print("=== ANALISIS BIVARIAT: UJI CHI-SQUARE ===")

# Daftar variabel independen yang diuji terhadap treatment_bin
support_vars = [
    "benefits", "care_options", "wellness_program",
    "seek_help", "anonymity", "leave"
]

stigma_vars = [
    "mental_health_consequence", "phys_health_consequence",
    "mental_health_interview", "phys_health_interview",
    "mental_vs_physical", "coworkers", "supervisor",
    "obs_consequence"
]

all_test_vars = support_vars + stigma_vars

hasil_chi = []  # untuk merangkum hasil

for var in all_test_vars:
    print("\n====================================================")
    print(f"Uji Chi-Square antara '{var}' dan 'treatment_bin'")
    
    # Crosstab
    ct = pd.crosstab(df_model_clean[var], df_model_clean["treatment_bin"])
    print("\nTabel Kontingensi:")
    display(ct)
    
    # Uji chi-square
    chi2, p, dof, expected = chi2_contingency(ct)
    
    print(f"Chi-square statistic: {chi2:.4f}")
    print(f"p-value: {p:.4f}")
    print(f"Degrees of freedom: {dof}")
    
    # Kesimpulan
    if p < 0.05:
        print("Kesimpulan: Ada hubungan signifikan (p < 0.05).")
    else:
        print("Kesimpulan: Tidak ada hubungan signifikan (p ≥ 0.05).")
    
    # Simpan ke ringkasan
    hasil_chi.append([var, chi2, p, dof])

# Ringkasan hasil chi-square
hasil_chi_df = pd.DataFrame(hasil_chi, columns=["Variabel", "Chi2", "p-value", "df"])
print("\n====================================================")
print("RINGKASAN HASIL UJI CHI-SQUARE")
display(hasil_chi_df.sort_values("p-value"))
```

```python
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd

# 1. Tentukan variabel independen yang akan dimasukkan ke model
support_vars = [
    "benefits", "care_options", "wellness_program",
    "seek_help", "anonymity", "leave"
]

stigma_vars = [
    "mental_health_consequence", "phys_health_consequence",
    "mental_health_interview", "phys_health_interview",
    "mental_vs_physical", "coworkers", "supervisor",
    "obs_consequence"
]

control_vars = [
    "Age", "Gender_simple", "no_employees", "remote_work"
]

all_X = support_vars + stigma_vars + control_vars
```

```python
# 2. Buat formula regresi logistik
formula = "treatment_bin ~ Age + C(Gender_simple) + C(no_employees) + C(remote_work)"

# Tambahkan dukungan perusahaan
for v in support_vars:
    formula += f" + C({v})"

# Tambahkan stigma
for v in stigma_vars:
    formula += f" + C({v})"

print("Formula regresi logistik:")
print(formula)
```

```python
# 3. Fit model regresi logistik
logit_model = smf.logit(formula=formula, data=df_model_clean).fit()

print("\n=== HASIL REGRESI LOGISTIK ===")
print(logit_model.summary())
```

```python
# 4. Hitung Odds Ratio
params = logit_model.params
conf = logit_model.conf_int()
odds_ratio = np.exp(params)
ci_lower = np.exp(conf[0])
ci_upper = np.exp(conf[1])

odds_table = pd.DataFrame({
    "Coef": params,
    "Odds Ratio": odds_ratio,
    "CI Lower": ci_lower,
    "CI Upper": ci_upper
})

print("\n=== TABEL ODDS RATIO ===")
display(odds_table)
```
